```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import os
import time
from selenium.webdriver.chrome.options import Options
```

```python
browser = webdriver.Chrome()
time.sleep(3)
browser.maximize_window()
time.sleep(3)
browser.get("https://www.thegioididong.com/")
element=browser.find_element_by_xpath("/html/body/header/div[1]/nav/a[2]")
element.click()
time.sleep(2)
data={"name":[],"price":[]}
for i in range(1,26):
    back=0
    try:
        element1=browser.find_element_by_xpath("/html/body/section/ul[2]/li[{0}]/a".format(i))
    except:
        pass
    try:
        element1.click()
    except:
        pass
    name=""
    price=""
    try:
        name=browser.find_element_by_xpath("/html/body/section/div[1]/h1")
    except:
        browser.back()
        back=back+1
    try:
        price=browser.find_element_by_xpath("""//*[@id="normalproduct"]/aside[2]/div[1]/strong""")
    except:
        if back==0:
            browser.back()
            back=back+1
        pass
    try:
        data["name"].append(name.text)
    except:
        pass
    try:
        data["price"].append(price.text)
    except:
        pass
    if back==0:
        browser.back()
    time.sleep(2)
```

```python
df = pd.DataFrame(data)
```
